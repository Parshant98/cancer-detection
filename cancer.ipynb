```python
x=df.drop(['Target'],axis=1)
y=df['Target']
```

```python
y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=True)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_trains=sc.fit_transform(x_train)
x_tests=sc.fit_transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
from sklearn.svm import SVC
```

```python
svmmodel=SVC()
```

```python
svmmodel.fit(x_trains,y_train)
```

```python
y_pred=svmmodel.predict(x_tests)
```

```python
print(accuracy_score(y_test,y_pred)*100,'%')
```

```python
svmmodel.fit(x_train,y_train)
y_predsv=svmmodel.predict(x_test)
print(accuracy_score(y_test,y_predsv))
```

```python
lrmodel=LogisticRegression()
lrmodel.fit(x_train,y_train)
y_predlr=lrmodel.predict(x_test)
print(accuracy_score(y_test,y_predlr))
```

# Randomized serarch

```python
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ] 
}
```

# xgboost with best parameter

```python
xgbestparm=XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
       colsample_bynode=1, colsample_bytree=0.3, gamma=0.0,
       learning_rate=0.2, max_delta_step=0, max_depth=25,
       min_child_weight=1, missing=None, n_estimators=100, n_jobs=1,
       nthread=None, objective='binary:logistic', random_state=0,
       reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
       silent=None, subsample=1, verbosity=1)
xgbestparm.fit(x_train,y_train)
xgbp_predict=xgbestparm.predict(x_test)
print(accuracy_score(y_test,xgbp_predict))
```
